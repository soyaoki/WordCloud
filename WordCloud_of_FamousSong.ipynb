```python
import pandas as pd
df = pd.read_csv('top_200_weekly.csv')
df.head()
```

## Gather lyrics

```python
import requests
from bs4 import BeautifulSoup

url = 'https://genius.com/' + df['Artist'][0].replace(" ","-") + '-' + df['Track Name'][0].replace(" ","-") + '-lyrics'
url
```

```python
res = requests.get(url)
```

```python
res
```

```python
soup = BeautifulSoup(res.content)
```

```python
for div in soup.findAll('div', attrs = {'class': 'lyrics'}):
    print(div.text.strip().split("\n"))
    text = div.text.strip().split("\n")
```

```python
text = " ".join(text).replace(",", " ")
text
```

```python
text = text.replace("[Verse 1]", " ")
text = text.replace("[Verse 2]", " ")
text = text.replace("[Chorus]", " ")
text = text.replace("[Pre-Chorus]", " ")
text = text.replace("[Bridge]", " ")
text = text.replace("[Intro]", " ")
text = text.replace("[Outro]", " ")
text = text.replace("Oh", " ")
text = text.replace("oh", " ")
text = text.replace("Ah", " ")
text = text.replace("ah", " ")
text = text.replace("Yeah", " ")
text = text.replace("yeah", " ")
text = text.replace("\\u", " ")
text = text.replace("\\", " ")
text
```

## Making WordCloud

```python
from wordcloud import WordCloud
wordcloud = WordCloud(background_color="white",
    width=800,height=600).generate(text)

wordcloud.to_file("wordcloud_test.png")
```

## All processes in loop

```python
%%time
from time import sleep
text_all = ""

for i in range(df['Artist'].shape[0]):
    url = 'https://genius.com/' + df['Artist'][i].replace(" ","-") + '-' + df['Track Name'][i].replace(" ","-") + '-lyrics'
    res = requests.get(url)
    soup = BeautifulSoup(res.content)
    for div in soup.findAll('div', attrs = {'class': 'lyrics'}):
        text = div.text.strip().split("\n")
        text = " ".join(text).replace(",", " ")
        text = text.replace("[Verse 1]", " ")
        text = text.replace("[Verse 2]", " ")
        text = text.replace("[Chorus]", " ")
        text = text.replace("[Pre-Chorus]", " ")
        text = text.replace("[Bridge]", " ")
        text = text.replace("[Intro]", " ")
        text = text.replace("[Outro]", " ")
        text = text.replace("Oh", " ")
        text = text.replace("oh", " ")
        text = text.replace("Yeah", " ")
        text = text.replace("yeah", " ")
        text = text.replace("\\u", " ")
        text = text.replace("\\", " ")
        text_all = text_all + " " + text
        sleep(1)
    if(i%(df['Artist'].shape[0]/10)==0):
        print("■")
```

```python
text_all
```

```python
wordcloud = WordCloud(background_color="white",
    width=800,height=600).generate(text_all)

wordcloud.to_file("wordcloud.png")
```
